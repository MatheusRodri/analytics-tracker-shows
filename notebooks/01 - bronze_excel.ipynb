```python
# Libs imports
from pyspark.sql.functions import regexp_replace,current_timestamp, input_file_name,col, dayofweek,when, date_format,hour,from_utc_timestamp, concat,year,month,lit,contains,lower, concat_ws,to_date,  sum as _sum
import datetime
import pandas as pd
import time
```

```python
PATH_ORIGIN = "/Volumes/sandbox_prd/raw_layer/files/tracker-shows/movies and series.xlsx"
BRONZE_NAME_TABLE = "sandbox_prd.bronze_layer.movies_and_series"
```

```python
# Read the excel
df_input = pd.read_excel(PATH_ORIGIN)
```

```python
#Add types
df_input['Name (PT-BR)'] = df_input['Name (PT-BR)'].astype(str)
df_input['Name (English)'] = df_input['Name (English)'].astype(str)
df_input['Order / Chronology'] = df_input['Order / Chronology'].astype(str)
df_input['Producer / Universe'] = df_input['Producer / Universe'].astype(str)
df_input['Type'] = df_input['Type'].astype(str)
df_input['Where'] = df_input['Where'].astype(str)
df_input['Status'] = df_input['Status'].astype(str)
```

```python

# Rename column
df_input = df_input.rename(
    columns={
        'Name (PT-BR)': 'name_pt_br', 
        'Name (English)': 'name_en', 
        'Order / Chronology': 'order_chronology', 
        'Producer / Universe': 'producer_universe',
        'Type': 'type', 
        'Where': 'where', 
        'When add': 'when_add', 
        'In progress': 'in_progress', 
        'Status': 'status', 
        'When Watched': 'when_watched'
        }
    )
```

```python
# Save bronze table
df_input = spark.createDataFrame(df_input)
df_input.write.format("delta").mode("overwrite").saveAsTable(BRONZE_NAME_TABLE)
```
