```python
SILVER_NAME_TABLE = "sandbox_prd.silver_layer.movies_and_series"
```

```python
%sql
CREATE OR REPLACE TEMPORARY VIEW movies_and_series_full AS
SELECT 
a.*,
b.*
FROM sandbox_prd.bronze_layer.movies_and_series AS a
LEFT JOIN sandbox_prd.bronze_layer.movies_and_series_api AS b
ON a.name_en = b.title_1
     
```

```python
df_silver_full = spark.sql("select * from movies_and_series_full")
```

```python

remove_columns = ['title','title_1']
df_silver = df_silver_full.drop(*remove_columns)
```

```python

df_silver_rename = df_silver\
    .withColumnRenamed('name_pt_br','nm_show_pt_br')\
    .withColumnRenamed('name_en','nm_show_en')\
    .withColumnRenamed('order_chronology','nr_order_chronology')\
    .withColumnRenamed('producer_universe','nm_producer_universe')\
    .withColumnRenamed('type','nm_type')\
    .withColumnRenamed('where','nm_where')\
    .withColumnRenamed('when_add','dt_when_add')\
    .withColumnRenamed('in_progress','st_in_progress')\
    .withColumnRenamed('status','st_status')\
    .withColumnRenamed('when_watched','dt_when_watched')\
    .withColumnRenamed('country','nm_country')\
    .withColumnRenamed('genre','nm_genre')\
    .withColumnRenamed('imdbRating','nr_imdbRating')\
    .withColumnRenamed('imdbVotes','nr_imdbVotes')\
    .withColumnRenamed('metascore','nr_metascore')\
    .withColumnRenamed('poster','link_poster')\
    .withColumnRenamed('rated','nm_rated')\
    .withColumnRenamed('released','dt_released')\
    .withColumnRenamed('runtime','nu_runtime')\
    .withColumnRenamed('totalSeasons','nu_totalSeasons')\
    .withColumnRenamed('year','dt_year')
```

```python

df_silver_rename.write.format("delta").mode("overwrite").saveAsTable(SILVER_NAME_TABLE)
```

```python
df_silver_rename.write.mode("overwrite").option("header", True).csv("/Volumes/sandbox_prd/silver_layer/files/tracker-shows/")
```
